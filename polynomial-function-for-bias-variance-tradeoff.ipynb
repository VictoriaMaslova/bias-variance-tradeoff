```python
import numpy as np
import matplotlib.pyplot as plt

#Creating our dataset
t = np.linspace(0,1,12)
h = np.array([1.5, 1.6, 1.7, 2.371, 2.35, 2.25, 2.3, 1.88, 1.6, 1.79, 1.07, 0.21])
plt.scatter(t,h)
```

```python
# Fitting using a linear equation (degree=1) to show high Bias
p1 = np.polyfit(t,h,1)
plt.scatter(t,h)
plt.plot(t, np.polyval(p1,t), c ='r')
plt.xlabel('t(s)')
plt.ylabel('h(m)')
plt.title('Model is underfit')
plt.legend()
```

```python
# Fitting using a linear equation (degree=10) to show high variance
p2 = np.polyfit(t,h,10)
plt.scatter(t,h)
plt.plot(t, np.polyval(p2,t), c ='r')
plt.xlabel('t(s)')
plt.ylabel('h(m)')
plt.title('Model is overfit')
plt.legend()
```

```python
#Fitting usuing a Quadratic equation( degree = 2) Bias-Variance Tradeoff
p3 = np.polyfit(t,h,2)
plt.scatter(t,h)
plt.plot(t, np.polyval(p3,t), c ='r')
plt.xlabel('t(s)')
plt.ylabel('h(m)')
plt.title('Quadratic Model is what we needed')
plt.legend()
```
