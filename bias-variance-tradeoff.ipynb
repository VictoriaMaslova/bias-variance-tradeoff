```python
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import pandas as pd
from sklearn.linear_model import LinearRegression
from mlxtend.evaluate import bias_variance_decomp
import numpy as np
import datetime
data = pd.read_csv('../input/szeged-weather/weatherHistory.csv')
data
```

```python
#Deleting unnecessary columns

data = data.drop(['Formatted Date','Daily Summary','Pressure (millibars)', 'Wind Speed (km/h)', 'Humidity', 'Loud Cover' ], axis = 1)
data
```

```python
#converting caterical data to numerical ones
label_encoder=preprocessing.LabelEncoder()
data['Summary'] = label_encoder.fit_transform(data['Summary'])
data['Summary'].unique()
data
```

```python
data['Precip Type']=data['Precip Type'].replace(['rain', 'snow'], [0,1])
data
```

```python
#Checking for NaN values in data
data.isna().sum()
```

```python
#Replacing missing values with ‘0’.
data['Precip Type']=data['Precip Type'].replace(np.nan,0)
data.isna().sum()
```

```python
#Let’s drop the prediction column from our dataset.
y = data['Apparent Temperature (C)']
data = data.drop('Apparent Temperature (C)', axis=1)
data
```

```python
y
```

```python
#Splitting the dataset into training and testing data and fitting our model to it.
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size = 0.2, random_state=0)

```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
#Finding variance using np
Variance = np.var(pred)
Variance
```

```python
#Using squared mean error to find bias
SSE = np.mean((np.mean(pred)-y)**2)
Bias = SSE - Variance
Bias
```

```python
# The mean square error
print("Residual sum of squares: %.2f" % SSE)

print("Bias: {Bias}".format(Bias=Bias))
print("Variance: {Variance}".format(Variance=Variance))
```
